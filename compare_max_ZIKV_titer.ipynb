```python
# Compare saliva samples

import pandas as pd
import scipy.stats as stats

ma66 = pd.read_table("ma66_sal_titers.txt", delim_whitespace=True, header=None, na_values='#VALUE!')
ma66_ave_data = ma66.copy()
ma66[2] = ma66_ave_data.mean(numeric_only=True, axis=1)
ma66_array = ma66[2].dropna().values

ma66_mean = ma66[[2]].mean()
ma66_med = ma66[[2]].median()
print('ma66',ma66_mean)
print(ma66_med)

pr = pd.read_table("PR_sal_titers.txt", delim_whitespace=True, header=None, na_values='#VALUE!')
pr_ave_data = pr.copy()
pr[2] = pr_ave_data.mean(numeric_only=True, axis=1)
pr_array = pr[2].dropna().values
pr_mean = pr[[2]].mean()
pr_med = pr[[2]].median()
print('pr',pr_mean)
print(pr_med)

print(pr_array, pr_array.mean())

t_stat, p_val = stats.ttest_ind(pr_array, ma66_array, equal_var=False)
print('pr',pr_array.mean(),'ma66',ma66_array.mean(),ma66_array.std())
print(t_stat)
print(p_val)

Wilcox_stat, p_val = stats.ttest_ind(pr_array, ma66_array)
print(Wilcox_stat)
print(p_val)
```

```python
# Compare LW samples

import pandas as pd
import scipy.stats as stats

ma66 = pd.read_table("ma66_lw_titers.txt", delim_whitespace=True, header=0, na_values='#VALUE!')
ma66_ave_data = ma66.copy()
ma66[2] = ma66_ave_data.mean(numeric_only=True, axis=1)
ma66_array = ma66[2].dropna().values
ma66_mean = ma66[[2]].mean()
ma66_med = ma66[[2]].median()
print('ma66',ma66_mean)

pr = pd.read_table("pr_lw_titers.txt", delim_whitespace=True, header=0, na_values='#VALUE!')
pr_ave_data = pr.copy()
pr[2] = pr_ave_data.mean(numeric_only=True, axis=1)
pr_array = pr[2].dropna().values
#pr_mean = pr[[2]].mean()
#pr_med = pr[[2]].median()
#print('pr',pr_mean)
#print(pr)

t_stat, p_val = stats.ttest_ind(pr_array, ma66_array, equal_var=False)
print('pr',pr_array.mean(),pr_array.std(),'ma66',ma66_array.mean(),ma66_array.std())
print(t_stat)
print(p_val)

print(len(pr_array), len(ma66_array))
Wilcox_stat, p_val = stats.ranksums(pr_array, ma66_array)
print(Wilcox_stat)
print(p_val)

y = pr[2].dropna()
x = np.random.normal(1, 0.04, size=len(y))
plt.plot(x, y, 'r.', alpha=0.8)
y = ma66[2].dropna()
x = np.random.normal(2, 0.04, size=len(y))
plt.plot(x, y, 'r.', alpha=0.8)

# remove low infected samples (<10^6)
ma66_lw_H = ma66.loc[ma66[2]>6]
ma66_lw_H_array = ma66_lw_H[2].dropna().values
pr_lw_H = pr.loc[pr[2]>6]
pr_lw_H_array = pr_lw_H[2].dropna().values

t_stat, p_val = stats.ttest_ind(pr_lw_H_array, ma66_lw_H_array, equal_var=False)
print(t_stat)
print(p_val)

print(len(pr_lw_H_array), len(ma66_lw_H_array))
Wilcox_stat, p_val = stats.ranksums(pr_lw_H_array, ma66_lw_H_array)
print(Wilcox_stat)
print(p_val)

print('pr',pr_lw_H_array.mean(),pr_lw_H_array.std(),'ma66',ma66_lw_H_array.mean(),ma66_lw_H_array.std())
plt.axhline(pr_lw_H_array.mean(),linestyle='--')
plt.axhline(ma66_lw_H_array.mean(),linestyle='--',color='k')
```
